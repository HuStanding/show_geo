```python
from pyecharts.faker import Faker
from pyecharts import options as opts
from pyecharts.charts import Geo
from pyecharts.globals import ChartType, SymbolType


def geo_base() -> Geo:
    c = (
        Geo()
        .add_schema(maptype="china")
        .add("geo", [list(z) for z in zip(Faker.provinces, Faker.values())])
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            visualmap_opts=opts.VisualMapOpts(),
            title_opts=opts.TitleOpts(title="Geo-基本示例"),
        )
    )
    return c
geo_base().render_notebook()
```

```python
def geo_guangdong() -> Geo:
    c = (
        Geo()
        .add_schema(maptype="广东")
        .add(
            "geo",
            [list(z) for z in zip(Faker.guangdong_city, Faker.values())],
            type_=ChartType.HEATMAP,
        )
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(
            visualmap_opts=opts.VisualMapOpts(),
            title_opts=opts.TitleOpts(title="Geo-广东地图"),
        )
    )
    return c
geo_guangdong().render_notebook()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Geo
from pyecharts.datasets import register_url

# 如果出现 ssl 异常的话加上下面两行代码
# import ssl
# ssl._create_default_https_context = ssl._create_unverified_context

def geo_echart_china_js() -> Geo:
    register_url("https://echarts-maps.github.io/echarts-china-counties-js/")

    geo = (
        Geo()
        .add_schema(maptype="海淀区")
        .set_global_opts(title_opts=opts.TitleOpts(title="海淀区"))
    )
    return geo
geo_echart_china_js().render_notebook()
```
